```python
import pandas as pd
import numpy as np

df = pd.read_csv('Mall_Customers.csv')
```

```python
df.head
```

```python
df.columns =(df.columns.str.strip().str.lower().str.replace(r"[()$]","",regex = True).str.replace("-","_").str.replace("","_"))
```

```python
df.columns
```

```python
df.info()
```

```python
x = df[['_a_n_n_u_a_l_ _i_n_c_o_m_e_ _k_','_s_p_e_n_d_i_n_g_ _s_c_o_r_e_ _1___1_0_0_']]
x.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(x['_a_n_n_u_a_l_ _i_n_c_o_m_e_ _k_'],x['_s_p_e_n_d_i_n_g_ _s_c_o_r_e_ _1___1_0_0_'])
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 5, random_state=42, n_init='auto')
labels = kmeans.fit_predict(x)
```

```python
labels[:10]
```

```python
df["cluster"] = labels
df.head()
```

```python
plt.scatter(
    x['_a_n_n_u_a_l_ _i_n_c_o_m_e_ _k_'],
    x['_s_p_e_n_d_i_n_g_ _s_c_o_r_e_ _1___1_0_0_'],
    c = labels
)


plt.xlabel('annual income in k')
plt.ylabel('spending score 1-100')
plt.title('customer segements')
plt.show()
```

```python
# elbow method
# it helps to find how many group to use
from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
  kmeans = KMeans(n_clusters = k, random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

#wcss - Within Cluster Sum of Squares
#Lower wcss - tightly cluster
#higher wcss - messy cluster_optics_dbscan

#we will store this for each kmeans
#each loop - new model - different number of clusters

# in each loop k increases ,wcss decreases and improvement slows down after certain point
```

```python
plt.plot(range(1,11),wcss)
```

```python
# applying PCA here

# Principal Compound Analysis - PCA

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x)
```

```python
plt.scatter(
    x_pca[:,0],
    x_pca[:,1],
    c=df['cluster']
)
```

```python
df.groupby('cluster')[['_a_n_n_u_a_l_ _i_n_c_o_m_e_ _k_','_s_p_e_n_d_i_n_g_ _s_c_o_r_e_ _1___1_0_0_']].mean()
```

```python
df['segement'] = df['cluster'].map({
    0:'mid income - mid spend',
    1:'high income - high spend',
    2:'low income - low spend',
    3:'high income -low spend',
    4:'low income - high spend'
})
```

```python
df.head()
```

```python
plt.figure(figsize=(10,10))
for segment_name,cluster_id in df[['segement','cluster']].drop_duplicates().values:
  subset=df[df['cluster']==cluster_id]

  plt.scatter(
      subset['_a_n_n_u_a_l_ _i_n_c_o_m_e_ _k_'],
      subset['_s_p_e_n_d_i_n_g_ _s_c_o_r_e_ _1___1_0_0_'],
      label=segment_name
  )

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segments')
plt.legend(title='Customer Segment')
plt.show()
```
